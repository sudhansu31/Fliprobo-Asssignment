# Scrapping saree images from Amazon site

```python
driver=webdriver.Chrome(r"C:\Users\Admin\Downloads\chromedriver.exe")
url="https://www.amazon.in/s?k=saree&page=2&qid=1637930860&ref=sr_pg_2"
driver.get(url)

page=8
p=1
img_url=[]
c=1
for i in range(1,page):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    pro_ele=driver.find_elements_by_class_name("s-expand-height.s-include-content-margin.s-latency-cf-section")
    print(len(pro_ele))
    for pro in pro_ele:
        print(c)
        try:
            pro.click()
            time.sleep(3)
            chwd=driver.window_handles
            driver.switch_to.window(chwd[1])
            time.sleep(2)
            try:
                image=driver.find_element_by_xpath('//div[@id="imgTagWrapperId"]')
                print(image)
                image.click()
            except:
                pass
            time.sleep(3)
            ima=driver.find_element_by_xpath('//*[@id="ivLargeImage"]/img')
            print(ima.get_attribute('src'))
            img_url.append(ima.get_attribute('src'))
            time.sleep(1)
            driver.close()
            driver.switch_to.window(chwd[0])
        
        except:
            pass
            
        
        c+=1
        time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    try:
        next_buttom=driver.find_element_by_class_name('a-last')
        next_buttom.click()
        time.sleep(2)
        print(p)
        p+=1
    except:
        break
       
```

```python
len(img_url)
```

```python
import pandas as pd
```

```python
img_url=pd.DataFrame(img_url)
```

```python
img_url.to_csv("saree_img_url")
```

# Downloading training images to specified folder

```python
os.chdir('D:/Dataset1/Training Dataset')
baseDir=os.getcwd()
```

```python
for i, url in enumerate(training_image_url["0"]):
    file_name = f"{i:150}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        
        file_path = os.path.join(baseDir, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
        print(f"ERROR - COULD NOT SAVE {url} - {e}")
```
