```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException
```

```python
driver=webdriver.Chrome(r"C:\Users\Admin\Downloads\chromedriver.exe")
url="https://www.amazon.in/s?i=electronics&rh=n%3A1389401031&fs=true&page=2&qid=1636740502&ref=sr_pg_1"
driver.get(url)
Review=[]
Rating=[]

page=20

for p in range(7,page+1):
    n=1
    print("page:",p)
    product_link=driver.find_elements_by_xpath('//div[@class="a-row a-size-small"]')
    print("Total number of product:",len(product_link))
    for product in product_link:
        print("Product no.:",n)
        try:
            
            product.click()
            time.sleep(2)
            chwd=driver.window_handles
            driver.switch_to.window(chwd[1])
            try:
                review_ele=driver.find_element_by_xpath('//*[@id="histogramTable"]/tbody/tr[1]/td[1]/span[1]')
                review_ele.click()
                click_ele=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]')
                time.sleep(2)
                for i in click_ele:
                    try:
                        Review.append(i.text)
                        Rating.append("5Star")
                    except:
                        Review.append("NaN")
                        Rating.append("NaN")
                
                print("Rating 5 text append")
                
            except:
                Print("Rating 5 star not fetched")
            driver.back()
            time.sleep(1)
            
            try:
                review_ele=driver.find_element_by_xpath('//*[@id="histogramTable"]/tbody/tr[2]/td[1]/span[1]')
                review_ele.click()
                click_ele=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]')
                time.sleep(2)
                for i in click_ele:
                    try:
                        Review.append(i.text)
                        Rating.append("4Star")
                    except:
                        Review.append("NaN")
                        Rating.append("NaN")
                        
                print("Rating 4 text append")
                
            except:
                Print("Rating 4 star not fetched")
            driver.back()
            time.sleep(1)
            
            try:
                review_ele=driver.find_element_by_xpath('//*[@id="histogramTable"]/tbody/tr[3]/td[1]/span[1]')
                review_ele.click()
                click_ele=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]')
                time.sleep(2)
                for i in click_ele:
                    try:
                        Review.append(i.text)
                        Rating.append("3Star")
                    except:
                        Review.append("NaN")
                        Rating.append("NaN")
                        
                print("Rating 3 text append")
                
            except:
                Print("Rating 3 star not fetched")
            driver.back()
            time.sleep(1)
            
            try:
                review_ele=driver.find_element_by_xpath('//*[@id="histogramTable"]/tbody/tr[4]/td[1]/span[1]')
                review_ele.click()
                click_ele=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]')
                time.sleep(2)
                for i in click_ele:
                    try:
                        Review.append(i.text)
                        Rating.append("2Star")
                    except:
                        Review.append("NaN")
                        Rating.append("NaN")
                        
                
                print("Rating 2 text append")
                
            except:
                Print("Rating 2 star not fetched")
            driver.back()
            time.sleep(1)
            
            try:
                review_ele=driver.find_element_by_xpath('//*[@id="histogramTable"]/tbody/tr[5]/td[1]/span[1]')
                review_ele.click()
                click_ele=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]')
                time.sleep(2)
                for i in click_ele:
                    try:
                        Review.append(i.text)
                        Rating.append("1Star")
                    except:
                        Review.append("NaN")
                        Rating.append("NaN")                        
                
                print("Rating 1 text append")
                
            except:
                Print("Rating 1 star not fetched")
            driver.back()
            time.sleep(1)
            
        except:
            n+=1
            continue
        driver.close()
        driver.switch_to.window(chwd[0])
        time.sleep(2)
        n+=1
    
    button_ele=driver.find_elements_by_class_name("a-last")
    for j in button_ele:
        j.click()
    time.sleep(2)
```
