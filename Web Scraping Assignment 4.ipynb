**1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:**

A) Rank

B) Name

C) Artist

D) Upload date

E) Views


```python
driver=webdriver.Chrome(r"C:\Users\Admin\Downloads\chromedriver.exe")
url=" https://en.wikipedia.org/w/index.php?search=List+of+most-viewed+YouTube+videos%2F&title=Special%3ASearch&fulltext=1&ns0=1"
driver.get(url)

link_ele=driver.find_element_by_xpath('//a[@href="/wiki/List_of_most-viewed_YouTube_videos"]')
link_ele.click()

time.sleep(3)

before_Xpath = "//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr["
aftertd_Xpath = "]/td["
aftertr_Xpath = "]"


rows = len(driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr"))


columns = len(driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[1]/td"))

time.sleep(2)
Rank=[]
Name=[]
Artist=[]
date=[]
view=[]
for t_row in range(1, (rows + 1)):
    for t_column in range(1, (columns)):
        FinalXPath = before_Xpath + str(t_row) + aftertd_Xpath + str(t_column) + aftertr_Xpath
        cell_text = driver.find_element_by_xpath(FinalXPath).text
        if t_column==1:
            Rank.append(cell_text)
        if t_column==2:
            Name.append(cell_text)
        if t_column==3:
            Artist.append(cell_text)
        if t_column==4:
            view.append(cell_text+str(" Billions"))
        if t_column==5:
            date.append(cell_text)
            
        
print("Rank:",len(Rank),"Name:",len(Name),"Artist:",len(Artist),"Date:",len(date),"view:",len(view))

data={"Rank":Rank,
     "Name":Name,
     "Artist":Artist,
     "Upload Date":date,
     "View":view}

report=pd.DataFrame(data)
report
```

**2. Scrape the details team India’s international fixtures from bcci.tv.

C) Place

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.


```python
driver=webdriver.Chrome(r"C:\Users\Admin\Downloads\chromedriver.exe")
url="https://www.bcci.tv/"
driver.get(url)

drop_ele=driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')
drop_ele.click()

fix_ele=driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
fix_ele.click()

title_ele=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
match_title=[]
for i in title_ele:
    try:
        match_title.append(i.text)
    except:
        match_title.append("-")

series_ele=driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
series=[]
for i in series_ele:
    try:
        series.append(i.text)
    except:
        series.append("-")
place_ele=driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]')
place=[]
for i in place_ele:
    try:
        str2=i.text
        str2=str2.split("\n")
        place.append(str2[1])
    except:
        place.append("-")

date_ele=driver.find_elements_by_xpath('//div[@class="fixture__full-date"]')
date=[]
time=[]
for i in date_ele:
    try:
        t=i.text
        str1=t.replace("\n","/")
        date.append(str1[:-10])        
    except:
        date.append("-")
    try:
        time.append(str1[-9:])
    except:
        time.append("-")

data={"Match Title":match_title,
     "Series": series,
     "Place": place,
     "Date":date,
     "Time":time}
report=pd.DataFrame(data)
report
```

```python
driver=webdriver.Chrome(r"C:\Users\Admin\Downloads\chromedriver.exe")
url="https://archive.ics.uci.edu/"
driver.get(url)

link_ele=driver.find_element_by_xpath('//html/body/table[2]/tbody/tr/td/span/b/a')
link_ele.click()

rows=len(driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr'))
print("No. of Records:",rows)

before_Xpath = "/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["
after_Xpath = "]"

Name=[]
Type=[]
Task=[]
Attribute=[]
Instances=[]
Noattribute=[]
Year=[]
for row in range(2,(rows+1)):
    for col in range(1,8):
        if col == 1:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath+str("/table/tbody/tr/td[2]/p/b/a")
                name=driver.find_element_by_xpath(finalXpath).text
                Name.append(name)
            except:
                Name.append("-")
        if col == 2:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                type=driver.find_element_by_xpath(finalXpath).text
                Type.append(type)
            except:
                Type.append("-")
        if col == 3:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                task=driver.find_element_by_xpath(finalXpath).text
                Task.append(task)
            except:
                Task.append("-")
        if col == 4:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                attri=driver.find_element_by_xpath(finalXpath).text
                Attribute.append(attri)
            except:
                Attribute.append("-")
        if col == 5:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                ins=driver.find_element_by_xpath(finalXpath).text
                Instances.append(ins)
            except:
                Instances.append("-")
        if col == 6:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                nattri=driver.find_element_by_xpath(finalXpath).text
                Noattribute.append(nattri)
            except:
                Noattribute.append(nattri)
        if col == 7:
            try:
                finalXpath=before_Xpath+str(row)+after_Xpath+str("/td[")+str(col)+after_Xpath
                year=driver.find_element_by_xpath(finalXpath).text
                Year.append(year)   
            except:
                Year.append("-")

print("length of Dataset name:",len(Name))
print("length of Data type:",len(Type))
print("length of Task:",len(Task))
print("length of Attribute type:",len(Attribute))
print("length of No of instances:",len(Instances))
print("length of No of attribute:",len(Noattribute)) 
print("length of Year:",len(Year))

data={
    "Dataset name":Name,
    "Data type":Type,
    "Task":Task,
    "Attribute type":Attribute,
    "No of instances":Instances,
    "No of attribute":Noattribute,
    "Year":Year    
}
           
report=pd.DataFrame(data)   
report
```
